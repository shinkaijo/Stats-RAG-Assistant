```python
import requests
from bs4 import BeautifulSoup
from markdownify import markdownify as md  # ✨ 치트키 패키지 추가

url = "https://online.stat.psu.edu/stat501/lesson/12/12.1"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

main_content = soup.find('section', class_='section') 

if main_content:
    # 🎯 get_text() 대신 markdownify를 사용해서 HTML을 마크다운으로 예쁘게 변환!
    # strip=['a', 'img'] 옵션을 주면 쓸데없는 하이퍼링크나 이미지 태그도 다 지워줍니다.
    clean_text = md(str(main_content), strip=['a'])
    
    # 위아래 쓸데없는 빈 줄 정리
    clean_text = clean_text.strip()
    
    print("--- 🚀 마크다운 변환 완료! ---")
    print(clean_text[:6000]) 
else:
    print("본문 태그를 찾지 못했습니다.")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate

# ✨ 새롭게 추가된 LCEL 필수 모듈
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# 1. 텍스트를 어떻게 썰지 규칙(도마) 세팅
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,       # 한 조각당 500자 내외로 썰기
    chunk_overlap=200      # 문맥이 끊기지 않게 50자씩 겹치게 썰기
)

# 2. 크롤링한 텍스트(clean_text)를 규칙대로 썰기
# (clean_text 변수는 이전 코드에서 정상적으로 만들어졌다고 가정합니다)
chunks = text_splitter.split_text(clean_text)
print(f"총 {len(chunks)}개의 텍스트 조각으로 썰렸습니다!")

# 3. 썰어둔 텍스트를 '벡터(숫자)'로 변환해서 FAISS 창고에 넣기
# 🚨 주의: 기존 코드에 닫는 괄호 ')'가 빠져 있어서 수정했습니다!
openai_api_key = "" 
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

# 4. 마법의 순간: 조각(chunks)과 변환기(embeddings)를 넣어 FAISS DB 완성!
vector_db = FAISS.from_texts(chunks, embeddings)

# FAISS DB를 검색기(Retriever)로 변환
retriever = vector_db.as_retriever()

# ---------------------------------------------------------

# 1. 우리의 챗봇 두뇌(LLM) 불러오기 
# 🚨 팩트 체크: 'gpt-5.4-mini'는 아직 출시되지 않은 모델입니다! 
# 현재 가장 빠르고 가성비 좋은 최신 모델인 'gpt-4o-mini'로 변경했습니다.
llm = ChatOpenAI(model="gpt-4o-mini", openai_api_key="", temperature=0)

# 2. 프롬프트 템플릿 세팅 (작성해주신 내용 그대로 유지)
prompt_template = """
너는 정보통계학과 학생들을 위한 친절하고 전문적인 통계 어시스턴트야.
사용자의 질문에 답변할 때, 반드시 아래에 제공된 [검색된 전공 자료]만을 근거로 답변해줘.

지시사항:
1. 검색된 영어 원문을 통계학 초보자도 이해하기 쉽게 자연스러운 한국어로 번역하고 풀어서 설명해줘.
2. 수식이 있다면 LaTeX 문법($ 기호)을 사용해서 깔끔하게 작성해줘.
3. 원문에 표(Table)나 상관계수 같은 '구체적인 데이터 수치'가 있다면, 절대 생략하지 말고 Markdown 표 형식으로 원본 그대로 예쁘게 그려서 설명해줘.
4. 4. 💡 [시각 자료의 논리적 배치] 원문에 산점도 행렬(Matrix Plot) 같은 그래프 이미지 링크(`![plot](/...)`)가 있다면, **네가 해당 변수들의 관계나 시각적 결과를 설명하는 문장 바로 뒤에 자연스럽게 이미지를 삽입해서 근거로 제시해.** - (예시: "이 산점도를 보면 Weight와 BSA가 강하게 연관되어 있음을 알 수 있습니다." 바로 뒤에 이미지 삽입)
   - 이미지 출력 시 주소는 반드시 `![산점도](https://online.stat.psu.edu원래주소)` 형태로 조합해서 써.)
5. 가지고 있는 자료에서 검색이 불가능한 질문은 모른다고 답해줘. 네 뇌피셜은 절대 추가하지 마.

[검색된 전공 자료]:
{context}

질문: {input}
"""

prompt = PromptTemplate.from_template(prompt_template)

# 3. 마법의 조립: LCEL 문법으로 파이프라인 구축 (기존 체인 2개 통합)
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# [검색기에서 문서 뽑고 텍스트화] -> [프롬프트] -> [LLM] -> [텍스트만 추출]
rag_chain = (
    {"context": retriever | format_docs, "input": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

# --- 🚀 최종 테스트: 드디어 진짜 챗봇에게 물어보기! ---
user_question = "다중공선성이란 무엇인가요? 왜 문제가 되나요?"

# LCEL 방식을 쓰면 딕셔너리 없이 질문 텍스트를 바로 넣고, 깔끔한 텍스트 답변을 바로 받습니다.
response = rag_chain.invoke(user_question)

print("\n🤖 [GPT-4o-mini의 최종 답변]")
print(response)
```
