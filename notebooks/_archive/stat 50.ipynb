```python
import requests
from bs4 import BeautifulSoup
from markdownify import markdownify as md
import time

# STAT 504 기본 주소 세팅
base_course_url = "https://online.stat.psu.edu/stat504"
all_documents_504 = ""

# Lesson 1부터 13까지 순회
for lesson_num in range(1, 14):
    # 🎯 수정 포인트: 스크린샷 주소창 규칙에 맞춰 'Lesson01', 'Lesson02' 형태로 조립
    # str(lesson_num).zfill(2)는 1을 '01'로 만들어 줍니다.
    lesson_str = f"Lesson{str(lesson_num).zfill(2)}"
    main_url = f"{base_course_url}/{lesson_str}"
    
    print(f"\n🚀 === [STAT-504 {lesson_str}] 크롤링 시작 === 🚀")
    
    response = requests.get(main_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # 알맹이가 담긴 <main class="content"> 추출
        main_content = soup.find('main', class_='content')
        
        if main_content:
            # ✨ 이미지(img) src 절대 경로 개조 작업
            for img_tag in main_content.find_all('img'):
                src = img_tag.get('src')
                if src:
                    if not src.startswith('http'):
                        # 보정된 절대 주소 주입
                        img_tag['src'] = f"{base_course_url}/{src.lstrip('/')}"
            
            # 마크다운 변환 (표와 보정된 이미지 링크 보존)
            clean_md = md(str(main_content), strip=['a']).strip()
            
            all_documents_504 += f"\n\n### Course: STAT 504 | URL: {main_url}\n"
            all_documents_504 += clean_md
            
            print(f"✅ {lesson_str} 수집 완료 (이미지 경로 및 표 보존)")
    else:
        print(f"❌ {lesson_str} 접근 실패 (상태 코드: {response.status_code}) - URL: {main_url}")
        
    time.sleep(2) # 서버 과부하 방지용 휴식 시간

# 안전한 로컬 경로에 백업 저장
save_path = r"C:/Users/xaxa0/opensource_statchatbot/vectordb_crawling/stat504_full_data.txt"
with open(save_path, "w", encoding="utf-8") as f:
    f.write(all_documents_504)

print(f"\n🎉 STAT 504 수집 및 파싱 완료! 결과 저장: [{save_path}]")
```
