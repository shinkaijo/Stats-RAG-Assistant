```python
import requests
from bs4 import BeautifulSoup
from markdownify import markdownify as md
import time

base_url = "https://online.stat.psu.edu/stat501/lesson"
all_documents = ""

# 큰 단원 레슨 1부터 15까지 반복 (STAT 501 전체 분량)
for lesson_num in range(1, 16):
    print(f"\n🚀 === [Lesson {lesson_num}] 크롤링 시작 === 🚀")
    
    # 1. 큰 단원 메인 페이지 (예: /lesson/1) 수집
    main_url = f"{base_url}/{lesson_num}"
    response = requests.get(main_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # 🎯 수정 포인트: 알려주신 정확한 본문 태그 적용!
        main_content = soup.find('section', class_='section')
        
        if main_content:
            all_documents += f"\n\n### URL: {main_url}\n"
            # img 태그는 살려두고(strip=['a']), 위아래 공백은 깔끔하게 제거(.strip())
            all_documents += md(str(main_content), strip=['a']).strip()
            print(f"✅ 메인 페이지 저장 완료: {main_url}")
    
    time.sleep(1.5) # 디도스 방지 매너 타임

    # 2. 하위 단원 무한 반복 수집 (예: /lesson/1/1.1, 1.2 ...)
    sub_num = 1
    while True:
        sub_url = f"{base_url}/{lesson_num}/{lesson_num}.{sub_num}"
        response = requests.get(sub_url)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            
            # 🎯 수정 포인트: 하위 페이지에서도 똑같이 정확한 태그로 핀셋 추출!
            main_content = soup.find('section', class_='section')
            
            if main_content:
                all_documents += f"\n\n### URL: {sub_url}\n"
                all_documents += md(str(main_content), strip=['a']).strip()
                print(f"  ↪ ✅ 하위 페이지 저장 완료: {sub_url}")
            
            sub_num += 1    # 다음 하위 페이지 번호로 증가
            time.sleep(1.5)
            
        else:
            # 404 에러 등이 뜨면 해당 레슨의 하위 페이지가 끝난 것이므로 종료
            print(f"  ❌ 하위 페이지 없음. (Lesson {lesson_num} 종료)")
            break

print("\n🎉 모든 데이터 크롤링 및 병합 완벽하게 완료!")

# 💡 꿀팁: 크롤링한 방대한 데이터를 매번 다시 긁어오면 힘드니까 txt 파일로 저장해둡시다.
save_path = r"C:/Users/xaxa0/오픈소스 프로젝트/벡터디비 크롤링/stat501_full_data.txt"
with open(save_path, "w", encoding="utf-8") as f:
    f.write(all_documents)
print("💾 stat501_full_data.txt 파일로 저장 완료!")
```
